#### Defining KPIs:

##### As this dataset is about e-commerce orders in India, there are some common KPI's in this business:

##### Transactional KPI's

- **#Orders**: the amount of orders; here I use the distinct count because one order can have more than one SKU.
- **Sales**: the sum of the amount. It is in "Indian rupee" currency.
- **Average order value (AOV)**: It is defined as sales/#orders. It shows, on average, the value of a ticket or purchase.
- **#Units**: The amount of units sold.
- **#Units/Order**: It shows on average how many units of one sku are sold by order.

##### Operational KPI's

- **Fill Rate**: It measures how many of the total orders have been delivered. It is defined as: Valid Orders / (Valid Orders + Cancelled Orders)

##### Defined that, the analysis will be based on the behavior of these KPI's.

##### For the transactional KPI's, I will use only the valid orders, where the status is 'Shipped', and for the operational KPI's, I will use all the datasets.


```python
# Do a copy of the new dataframe
full_df = new_df.copy()
# Filter the new dataframe with only the rows with 'Status' = 'Shipped' to the transactional analysis
df = new_df[new_df['Status'] == 'Shipped']

print(full_df.shape)
print(df.shape)
```

```python
# Create a table to view #Orders and Sales by month
df_kpis = df.groupby(df['Date'].dt.month).agg({
    'Order ID': pd.Series.nunique,
    'Amount': 'sum'
}).reset_index()

df_kpis = df_kpis.rename(columns={'Order ID': '#Orders', 'Amount': 'Sales'})

# Calculate the Average Order Value (AOV)
df_kpis['AOV'] = df_kpis['Sales'] / df_kpis['#Orders']
df_kpis
```

```python
# Create the figure and axis
fig, ax1 = plt.subplots(figsize=(5, 3))

# Barplot for the orders
sns.barplot(x=df_kpis['Date'].astype(str), y=df_kpis['#Orders'], ax=ax1)

# Create a second axis for the AOV
ax2 = ax1.twinx()

# Line chart for the AOV, using the same axis values as before
sns.lineplot(x=df_kpis['Date'].astype(str), y=df_kpis['AOV'], ax=ax2, marker='o', color='red')

# Additional setup
ax1.set_xlabel('Month')
ax1.set_ylabel('# Orders')
ax2.set_ylabel('AOV', color='red')
ax1.set_title('Orders and AOV per Month')

# Adjust the scale of AOV
ax2.set_ylim(0, df_kpis['AOV'].max() + 5 * df_kpis['AOV'].std())

# Show the cart
plt.show()
```

##### Insight:

#### CROSS BASKET

##### As said before, one order can have more than one SKU, so it is important to analyze if there is any "trend" or "interaction" between SKUS. I mean, SKUS that tend to be bought together. I will do the analysis to determine if there are some SKUS that have this behavior and the same with categories.


##### SKU Cross Basket


```python
# Group by Order ID and make a list of all SKUS within that order
df_grouped_sku = df.groupby('Order ID')['SKU'].apply(lambda x: sorted(list(set(x)))).reset_index()
# Rename the columns
df_grouped_sku.columns = ['Order ID', 'SKU_LIST']
# Calulate the length of that list. Length = number of skus in that order
df_grouped_sku['SKU_LIST_LENGHT'] = df_grouped_sku['SKU_LIST'].apply(len)
# Sort the values by the lenght of the list
df_grouped_sku.sort_values('SKU_LIST_LENGHT',ascending=False, inplace=True)

# Do this because if I leave SKU_LIST as a list, it would not be possible to group by it later, so I change "," for " | "
df_grouped_sku['SKU_LIST'] = df_grouped_sku['SKU_LIST'].apply(lambda x: ' | '.join(x))
# Get the total amount of the order from the original dataset
df_grouped_sku['ORDER_AMOUNT'] = df['Amount'].groupby(df['Order ID']).transform('sum')

df_grouped_sku.reset_index(inplace=True, drop= True)

df_grouped_sku.head()
```

```python
# Create a table for getting the KPIs for the SKU_LIST for those orders with more than 1 SKU
df_sku_list_ranking = (
    df_grouped_sku[df_grouped_sku["SKU_LIST_LENGHT"] > 1]
    .groupby(["SKU_LIST"])
    .agg(
        {"Order ID": pd.Series.nunique,
        "ORDER_AMOUNT":'sum'}
        )
    .sort_values("Order ID", ascending=False)
    .reset_index()
)

df_sku_list_ranking = df_sku_list_ranking.rename(columns= {"Order ID":"#Orders" , "ORDER_AMOUNT":"Sales"})

df_sku_list_ranking["AOV"] = df_sku_list_ranking["Sales"]/df_sku_list_ranking["#Orders"]
# Get the top 10 ranking
df_sku_list_ranking.head(10)
```

```python
# Create the barplot
sns.barplot(x=df_sku_list_ranking['#Orders'].head(10), y=df_sku_list_ranking['SKU_LIST'].head(10), orient='h')

# set title and labels
plt.title('Top 10 SKU Combinations by #Orders')
plt.xlabel('#Orders')
plt.ylabel('SKU Combination')
plt.show()
```

```python
# Create the barplot
sns.barplot(x=df_sku_list_ranking['AOV'].head(10), y=df_sku_list_ranking['SKU_LIST'].head(10), orient='h')

# set title and labels
plt.title('Top 10 SKU Combinations by AOV')
plt.xlabel('AOV')
plt.ylabel('SKU Combination')
plt.show()
```

##### Categories Cross Basket


```python
# Group by Order ID and make a list of all Categories within that order
df_grouped_category = df.groupby('Order ID')['Category'].apply(lambda x: sorted(list(set(x)))).reset_index()
# Rename the columns
df_grouped_category.columns = ['Order ID', 'CATEGORY_LIST']
# Calulate the length of that list. Length = number of categories in that order
df_grouped_category['CATEGORY_LIST_LENGHT'] = df_grouped_category['CATEGORY_LIST'].apply(len)
# Sort the values by the lenght of the list
df_grouped_category.sort_values('CATEGORY_LIST_LENGHT',ascending=False, inplace=True)

# Do this because if I leave SKU_LIST as a list, it would not be possible to group by it later, so I change "," for " | "
df_grouped_category['CATEGORY_LIST'] = df_grouped_category['CATEGORY_LIST'].apply(lambda x: ' | '.join(x))
# Get the total amount of the order from the original dataset
df_grouped_category['ORDER_AMOUNT'] = df['Amount'].groupby(df['Order ID']).transform('sum')

df_grouped_category.reset_index(inplace=True, drop= True)

df_grouped_category.head()
```

```python
# Create a table for getting the KPIs for the CATEGORY_LIST for those orders with more than one category
df_category_list_ranking = (
    df_grouped_category[df_grouped_category["CATEGORY_LIST_LENGHT"] > 1]
    .groupby("CATEGORY_LIST")
    .agg(
        {"Order ID": pd.Series.nunique,
         "ORDER_AMOUNT": "sum"})
    .sort_values("Order ID", ascending=False)
    .reset_index()
)

df_category_list_ranking = df_category_list_ranking.rename(columns= {"Order ID":"#Orders" , "ORDER_AMOUNT":"Sales"})

df_category_list_ranking["AOV"] = df_category_list_ranking["Sales"]/df_category_list_ranking["#Orders"]

# Get the top 10 ranking
df_category_list_ranking.head(10)
```

```python
# Create the barplot
sns.barplot(x=df_category_list_ranking['#Orders'].head(10), y=df_category_list_ranking['CATEGORY_LIST'].head(10), orient='h')
# Set title and labels
plt.title('Top 10 Category Combinations by #Orders')
plt.xlabel('#Orders')
plt.ylabel('Category Combination')
plt.show()
```

```python
# Create the barplot
sns.barplot(x=df_category_list_ranking['AOV'].head(10), y=df_category_list_ranking['CATEGORY_LIST'].head(10), orient='h')
# Set title and labels
plt.title('Top 10 Category Combinations by AOV')
plt.xlabel('AOV')
plt.ylabel('Category Combination')
plt.show()
```
